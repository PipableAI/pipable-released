```python
try:
    connection = psycopg2.connect(
        host=host,
        user=username,
        password=password,
        database=database
    )

    cursor = connection.cursor()

    # write code to print all schema names
    print("Schema names:")
    cursor.execute("SELECT schema_name FROM information_schema.schemata;")
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    #write code to print all table names in the alyf schema of the database
    print("Table names in alyf schema:")
    cursor.execute("SELECT table_name FROM information_schema.tables WHERE table_schema = 'alyf';")
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    # write code to print all column names of a table and its data type
    print("Column names and data types of alyf_asvs:")
    cursor.execute("SELECT column_name, data_type FROM information_schema.columns WHERE table_name = 'alyf_asvs';")
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    # write code to print all column names of alyf_members and its data type
    print("Column names and data types of alyf_members:")
    cursor.execute("SELECT column_name, data_type FROM information_schema.columns WHERE table_name = 'alyf_members';")
    rows = cursor.fetchall()
    for row in rows:
        print(row)

    cursor.close()
    connection.close()

except (Exception, psycopg2.Error) as error:
    print("Error connecting to PostgreSQL:", error)
```
